```
from sklearn.model_selection import train_test_split

def form_content_to_split(label):
  filepath = "/content/"+label+"_pre.txt"
  file = open(filepath, "r")
  content = file.read()
  split_content = content.split(";")
  filtered_content = list(filter(lambda x: x != '', split_content))
  return filtered_content
  
def write_contents_to_file(content, label, filetype):
  file_to_write = open("/content/" + label + "_" + filetype + ".txt" , 'w')
  file_to_write.write(content)
  file_to_write.close()

def split_contents(content):
  train_set, test_set = train_test_split(content, test_size=0.39)
  return train_set, test_set

def form_final_content(label, unintended_label1, unintended_label2):
  labelled_content = ' 1\n'.join('{}'.format(k) for k in label) + " 1\n"
  unintended_content1 = ' -1\n'.join('{}'.format(k) for k in unintended_label1) + " -1\n"
  unintended_content2 = ' -1\n'.join('{}'.format(k) for k in unintended_label2) + " -1\n"
  final_content = labelled_content + unintended_content1 + unintended_content2
  return final_content
```

Get the content ready for Go

```
new_other_content_go = form_content_to_split("Go")
print(len(new_other_content_go))
```

Split whole filenames into train and test

```
train_set_go, test_set_go = split_contents(new_other_content_go)
print(train_set_go)
```

Get the content ready for Stop

```
new_other_content_stop = form_content_to_split("Stop")
print(len(new_other_content_stop))
```

Split whole filenames into train and test

```
train_set_stop, test_set_stop = split_contents(new_other_content_stop)
print(test_set_stop)
```

Get the content ready for Warn

```
new_other_content_warn = form_content_to_split("Warn")
print(len(new_other_content_warn))
```

Split whole filenames into train and test

```
train_set_warn, test_set_warn = split_contents(new_other_content_warn)
print(test_set_warn)
```

Get the content for Training-Imageset file for all labels

```
final_content_go_train = form_final_content(train_set_go, train_set_stop, train_set_warn)
final_content_stop_train = form_final_content(train_set_stop, train_set_go, train_set_warn)
final_content_warn_train = form_final_content(train_set_warn, train_set_stop, train_set_go)                            
```

Get the content for validation-imagesets for all labels

```
final_content_go_test = form_final_content(test_set_go, test_set_stop, test_set_warn)
final_content_stop_test = form_final_content(test_set_stop, test_set_go, test_set_warn)
final_content_warn_test = form_final_content(test_set_warn, test_set_stop, test_set_go)  
```

Write all the contents to file

```
write_contents_to_file(final_content_go_train, "Go", "train")
write_contents_to_file(final_content_stop_train, "Stop", "train")
write_contents_to_file(final_content_warn_train, "Warn", "train")
write_contents_to_file(final_content_go_test, "Go", "val")
write_contents_to_file(final_content_stop_test, "Stop", "val")
write_contents_to_file(final_content_warn_test, "Warn", "val")
```
